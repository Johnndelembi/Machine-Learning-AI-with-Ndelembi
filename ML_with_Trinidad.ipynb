## 1. Introduction to Machine Learning

### What does it mean by Machine Learning in simple terms...

**Machine learning** is like giving a computer the ability to learn from experience, just like we do! Here's the gist:

- We show the computer a bunch of examples: This could be data like pictures, text, numbers, or even sounds. Let's say you want a computer to identify different types of animals in pictures. You'd show it many pictures of cats, dogs, birds, and so on.

- The computer looks for patterns: It analyzes all this data to find similarities and differences between the examples. In the animal example, it might identify patterns in fur texture, ear shapes, or body types to distinguish cats from dogs.

- The computer makes predictions: Once it has learned from the examples, it can try to predict what something new is. For instance, if you show the computer a picture of an animal it hasn't seen before, based on what it learned from the training data, it might predict if it's a cat or a dog (or maybe something else entirely!).

**The key thing** is that the computer learns on its own, without needing someone to explicitly program it for every single situation. It gets better at making predictions the more data it sees.

**General definition**

Machine learning is like training a super smart friend. You show them examples (data) like pictures or numbers, and they learn to recognize patterns. This lets them make guesses (predictions) about new things they haven't seen before, like identifying animals in a photo. The more they learn, the better they get!

### Types of Machine Learning...

**Machine learning** encompasses **algorithms** that improve their performance on a specific task by **analyzing data**, **identifying patterns**, and **making data-driven predictions.** These algorithms can range from *(i)supervised learning,* where labeled data guides the model, to *(ii)unsupervised learning*, where the model finds hidden structures within unlabeled data. This allows machines to adapt and solve problems without explicit programming for every scenario.

### Application of Machine Learning
- Movie Recommendation on netflix
- Music streaming and personalized playlists
- Youtube videos algorithms

### The Machine Learning Workflow (In Layman terms)

- **Data, Data Everywhere:** The journey starts with data! This could be anything from pictures and text to numbers and even sounds. It's the raw material your machine learning friend will learn from.

- **Ask the Right Question:** What do you want your machine learning friend to learn?  Is it recognizing objects in images, predicting future sales, or something else entirely? Defining a clear goal helps guide the process.

- **Data Preparation:**  Imagine your friend's messy room before cleaning.  Data prep is like tidying up your data. You clean it, organize it, and get rid of anything irrelevant so your friend can learn from the most useful information.

- **Model Selection:** There are different machine learning methods (models) like decision trees or neural networks. Choosing the right model is like picking the best teaching tool for your friend.

- **Training Time!:** Now comes the learning part. You feed your prepared data to the chosen model. It analyzes the data, identifies patterns, and essentially learns how to perform the task you defined.

- **Testing, Testing, 1, 2, 3:** Just like any good student, your machine learning friend needs to be tested. You show it new data it hasn't seen before and see how well it performs its task (predictions).

- **Fine-Tuning (Optional):** If the results aren't perfect, don't worry! You can adjust the model or provide more data to help it improve its performance. It's like giving your friend extra practice problems.

- **Deployment (The Big Moment!):** Once your machine learning friend is well-trained and makes good predictions, it's time to use its newfound knowledge in the real world! This could be integrating it into an app, website, or any system that can benefit from its skills.

### Data Transformation: Feature scaling, normalization, encoding categorical variables

- Downloading and loading the necessary packages for this part

```python
#pip install sklearn, xgboost 
```

```python
#loading up sklearn package
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import confusion_matrix, r2_score, classification_report, mean_squared_error, mean_absolute_error, accuracy_score, precision_score, recall_score, f1_score
```

```python
# Label Encoding for categorical features
categorical_features = df[['Field Name', 'WELL_BORE_CODE', 'N_WELL_BORE_CODE', 'WellBore Name', 'FLOW_KIND', 'WELL_TYPE']]
le = LabelEncoder()
for feature in categorical_features:
  df[feature] = le.fit_transform(df[feature])
```

```python
df.head(3)
```

### Feature Engineering: Creating new features from existing ones

```python
# Separate features (x) and target variable (y)
x = df.drop({'Oil Production (stb/day)', 'Gas Volume (scf/day)', 'Water Production (stb/day)'}, axis=1)
y = df[['Oil Production (stb/day)', 'Gas Volume (scf/day)', 'Water Production (stb/day)']]
```

```python
# Feature scaling
scaler = StandardScaler() 
x = scaler.fit_transform(x)  # Scaling all independent variables/features
```

```python
# Prepare data for XGBoost (XGBoost requires numpy arrays)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Regression, Classification, Model Evalutation, Model Evaluation and Hyper-params tuning 

## Regression Problem (Using RandomForestRegressor model)

### Terms
**GridSearchCV is a hyperparameter tuning technique in machine learning that exhaustively searches over a specified parameter grid for the best combination of hyperparameters.**

- **estimator**: The machine learning model to be tuned (e.g., RandomForestClassifier, SVC).
- **param_grid:** A dictionary or list of dictionaries specifying the hyperparameter values to be explored.
- **scoring:** The evaluation metric used to compare different parameter combinations (e.g., 'accuracy', 'precision', 'recall', 'f1', 'neg_mean_squared_error').
- **cv:** The cross-validation strategy to use (e.g., integer for number of folds, a CV splitter object).
- **n_jobs:** The number of jobs to run in parallel (for faster computation).

**RandomizedSearchCV:** is another hyperparameter tuning technique that explores the parameter space more efficiently than GridSearchCV. Instead of exhaustively trying every combination of hyperparameters, it randomly samples a specified number of combinations.

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

random_forest = RandomForestRegressor()
params_grid = {
    "n_estimators": [100, 200],
    "max_depth": [5, 10, 15],
    #"learning_rate": [0.1, 0.2],
    "random_state": [42],
    #"tree": [5, 10, 15],
    
}

grid_search = GridSearchCV(random_forest, params_grid, cv=5, n_jobs=5, scoring="neg_mean_squared_error", return_train_score=True)
grid_search.fit(X_train, y_train)

#Scroing can either be >> 'accuracy','precision', 'recall', 'f1', 'neg_mean_squared_error' 
#NOTE: score values like accuracy, precision, f1-score and recall are for classification problem while, neg_mean_squared_error is for regression problem.
```

```python
best_random_forest = grid_search.best_estimator_

best_random_forest.score(X_test, y_test)
```

```python
print("score values of the best estimator obtained\n")
print(grid_search.best_params_)
```

### When to Use Which:

**GridSearchCV:** If you have a relatively small search space and want to explore all possible combinations.

**RandomizedSearchCV:** When dealing with a large number of hyperparameters or computational resources are limited.

# Regression Problem (Using XGRegressor model)

**XGBoost (eXtreme Gradient Boosting)** is a powerful machine learning model known for its exceptional performance in regression tasks. Here's why it stands out:

- Ensemble Power: XGBoost is an ensemble method, meaning it combines multiple weaker decision trees into a "forest" that makes stronger predictions. This technique helps reduce errors and improve accuracy compared to a single decision tree.

- Gradient Boosting: Unlike standard boosting, XGBoost uses a technique called gradient boosting. This focuses on improving the model's performance on past errors, leading to more precise predictions with each iteration.

- Regularization: XGBoost incorporates regularization, which helps prevent overfitting. Overfitting occurs when a model memorizes the training data too well and performs poorly on unseen data. Regularization techniques in XGBoost control model complexity, preventing this issue.

- Scalability and Efficiency: XGBoost is known for its efficiency in handling large datasets. It utilizes techniques like parallel processing and efficient memory usage, making it suitable for real-world applications with massive data volumes.

While not universally the "best" for every regression problem, XGBoost's combination of these factors makes it a strong contender:

**Accuracy:** Often achieves high accuracy on various regression tasks, especially compared to simpler models.

**Flexibility:** Can handle different types of data, including continuous and categorical features.

**Interpretability:** To some extent, XGBoost models offer insights into feature importance, helping understand which factors have the most impact on the predictions.

Read documentation of xgboost model for further understanding: >> https://xgboost.readthedocs.io/en/stable/

```python
# XGBoost specifically requires float data types
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
y_train = y_train.astype('float32')
y_test = y_test.astype('float32')
```

```python
from sklearn.model_selection import GridSearchCV

xgboost = XGBRegressor()
params_grid = {
    "n_estimators": [100, 200],
    "max_depth": [5, 10, 15],
    "learning_rate": [0.1, 0.2],
    #"random_state": [42],
    #"tree": [5, 10, 15],
    
}

grid_search = GridSearchCV(xgboost, params_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
best_xg_boost = grid_search.best_estimator_
best_xg_boost.score(X_test, y_test)
```
